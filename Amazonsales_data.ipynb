```python
item_type_revenue = data.groupby('Item Type')['Total Revenue'].sum()

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(item_type_revenue, labels=item_type_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Revenue Distribution by Item Type')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```
